#### Création d'un RDD contenant des nombres de 0 à 499. Notez bien le type d'objet manipulé (RDD -> Distribué)

```python
from pyspark import SparkContext
sc = SparkContext("local", "app")
```

```python
liste = []
for i in range(500):
  liste.append(i)
  
rdd1 = sc.parallelize(liste)
```

#### Affichage du contenu et de la taille de la RDD

```python
counts = rdd1.count()
print("Number of elements in RDD -> %i" % (counts))
coll = rdd1.collect()
print("Elements in RDD -> %s" % (coll))
```

#### Afficher le nombre de partition du RDD (le réduire à 2 puis le mettre à 4) 
#### Deux fonctions sont disponibles pour gérer le nombre de partitions : leur fonctionnement est différent, Essayez de comprendre en quoi!

```python
print("Number of partitions : ", rdd1.getNumPartitions())
# Regarde la donnée : print("Number of partitions : ", rdd1.partitionBy(2).getNumPartitions())

print("Number of partitions : ", rdd1.repartition(2).getNumPartitions()) # Shuffle des données entre les partitions donc prend du temps de calcul
print("Number of partitions : ", rdd1.coalesce(4).getNumPartitions()) # Ne peut que réduire le nombre de partition
```

#### Récupérer un échantillon de données (disons 100 données, prises au hasard). Deux méthodes existent, expliquer la différence.

```python
print("Sample : ", rdd1.sample(False, 0.2).count()) # RDD - mieux pour bcp de données
print("TakeSample : ", len(rdd1.takeSample(True, 100))) # Liste Python
```

#### Quelle est la différence entre Take et TakeOrdered ?

```python
# Take the first num elements of the RDD
rdd1.take(10) 
```

```python
# Get the N elements from an RDD ordered in ascending order or as specified by the optional key function
rdd1.takeOrdered(10,key=lambda x: -x)
```

#### Différence entre Map et FlatMap ? Ecrire deux transformations en utilisant chacune de ces fonctions

```python
# MAP : ajouter 1 à chaque élément du rdd
sorted(rdd1.map(lambda x: (x+1)).collect())
```

```python
sorted(rdd1.flatMap(lambda x: [x+1]).collect())
```

#### Liser le fichier enregistré précédemment (avec spark.sparkContext.textFile ET avec spark.sparkContext.wholeTextFiles). Quelle est la différence entre ces deux primitives ?

```python
# Takes an URI for the file and reads it as a collection of lines
data = sc.textFile('spark.csv')
print(data.collect())
```

```python
# Read a directory containing multiple small text files, and returns each of them as (filename, content) pairs.
# This is in contrast with textFile, which would return one record per line in each file
data = sc.wholeTextFiles('spark.csv')
print(data.collect())
```
